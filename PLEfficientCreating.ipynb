```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('E:\\Coding\\WebScrapingPractice\\PLStatAnalysis\\EC2.csv')

#This creates versions of the df that remove players
dfsca = df[(df['sca90'] != 0)]
dfgca = df[(df['gca90'] != 0)]

df.describe() # look at means for the gridlines.
```

```python
#SCATTERPLOT #1: Goal Creating Actions vs Touches (Per 90 mins)

othergca = dfgca[(dfgca['squad'] != 'West Ham')]
whugca = dfgca[(dfgca['squad'] == 'West Ham')]

plt.figure(figsize=(6, 4))
plt.scatter(whugca['gca90'], whugca['touchesp90'], color='maroon', marker='o', s=35, label='WHU') #The code that actually produces the scatterplot.
plt.scatter(othergca['gca90'], othergca['touchesp90'], color='black', marker='o', s=30, label = 'Other PL Team')
for i, row in whugca.iterrows(): #logic to itterate through the rows/datapoints and grab associated player name
    # Check if the squad is 'West Ham' (case-insensitive comparison) so only west ham players are selected for name printing.
     plt.text(row['gca90'], row['touchesp90'], row['player'], fontsize=5, ha='center', va='top', color = 'white')

plt.xlim(0, 1.2) #Set limits for each axis
plt.ylim(20, 110)

midx=0.34 #assign variables for the diff bins.
midy=51
plt.axvline(x=midx, color='black', linestyle='--', linewidth=1)
plt.axhline(y=midy, color='black', linestyle='--', linewidth=1)
plt.title('Creative Efficiency: Goal Creating Actions vs No. of Touches', color='white')
plt.xlabel('GCA p/90', color='white')
plt.ylabel('Touches p/90', color='white')
plt.gcf().set_facecolor('darkgrey') #color of the overall figure
plt.gca().set_facecolor('grey')#color of bkgrnd between the axis/on the graph
plt.grid(linestyle='--', alpha=0.5) #produces general gridlines



box_props = dict(boxstyle='round,pad=0.3', facecolor='white', edgecolor='black', alpha=0.5)
plt.text(midx * 2.5, midy * .78, 'Efficient\nCreators', horizontalalignment='center', verticalalignment='center', fontsize=4, bbox=box_props)
plt.text(midx * 2, midy * 1.78, 'High Vol.\nTouch Creators', horizontalalignment='center', verticalalignment='center', fontsize=4, bbox=box_props)
plt.text(midx * .3, midy * 1.65, 'High Vol Touch\nLess Creation', horizontalalignment='center', verticalalignment='center', fontsize=4, bbox=box_props)
plt.text(midx * 2.5, midy * .2, '(SAMPLE\nPL MF/AT ONLY\n5+ 90s)', horizontalalignment='center', verticalalignment='center', fontsize=4, bbox=box_props)
plt.text(midx * .2, midy * .48, 'Few Touches\nLittle Creation', horizontalalignment='center', verticalalignment='center', fontsize=4, bbox=box_props)

plt.text(midx * 1, midy * .4, 'Mean GCA p/90', horizontalalignment='center', verticalalignment='center', fontsize=3.5, bbox=box_props)
plt.text(midx * 3.5, midy * 1, 'Mean\nTouches p/90', horizontalalignment='center', verticalalignment='center', fontsize=3.5, bbox=box_props)

plt.savefig('GCA90vsTouches.png', dpi=300) # Increase DPI for higher resolution
```
